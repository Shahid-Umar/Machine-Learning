# Inverse encoding/ Inverse transform /Decoding categorical variables

Inverse encoding is the process of converting the encoded data back to its original form. This is done by using the `inverse_transform()` method. This method is available for the following encoders:

- `LabelEncoder`
- `OrdinalEncoder`
- `OneHotEncoder`

# **1. LABLE ENCODING**

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
# load the data
df = sns.load_dataset('titanic')
df.head()
```

```python
# impute missing values for age, embarked, embark_town and deck
df['age'].fillna(df['age'].median(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
# drop deck column
df.drop('deck', axis=1, inplace=True)
```

```python
# encode all categorical variables and object variables

le_sex = LabelEncoder()
le_embarked = LabelEncoder()
le_class = LabelEncoder()
le_who = LabelEncoder()
le_embark_town = LabelEncoder()
le_alive = LabelEncoder()

df['sex'] = le_sex.fit_transform(df['sex'])
df['embarked'] = le_embarked.fit_transform(df['embarked'])
df['class'] = le_class.fit_transform(df['class'])
df['who'] = le_who.fit_transform(df['who'])
df['embark_town'] = le_embark_town.fit_transform(df['embark_town'])
df['alive'] = le_alive.fit_transform(df['alive'])
df.head()
```

```python
# inverse transform the data

df['sex'] = le_sex.inverse_transform(df['sex'])
df['embarked'] = le_embarked.inverse_transform(df['embarked'])
df['class'] = le_class.inverse_transform(df['class'])
df['who'] = le_who.inverse_transform(df['who'])
df['embark_town'] = le_embark_town.inverse_transform(df['embark_town'])
df['alive'] = le_alive.inverse_transform(df['alive'])
df.head()
```

# **2. ONE HOT ENCODING**

- Method 1: Simple coding snippet of OneHotEncoding

```python
from sklearn.preprocessing import OneHotEncoder

df = sns.load_dataset('titanic')

cat_columns = ['sex','embarked']

encoder = OneHotEncoder(sparse_output=False) # for getting different type of output
encoded_df = pd.DataFrame(encoder.fit_transform(df[cat_columns]))

# concatenate the dataframes 
df = pd.concat([df, encoded_df], axis=1)
#df.drop(cat_columns, axis=1, inplace=True)
pd.set_option('display.max_columns', None)
df.head(10)
```

### How works One Hot Encoding?
- If we apply OneHotEncoding on column(s) then it creates seperate columns in dataset in respect of unique values in that variable.\
- In above example, 'sex' and 'embarked' columns 2 & 4 unqiue values respectively. So, output generate six extra columns at the end of dataset named 0,1,2,3,4,5

- Method 1: Using for loop code snippet of OneHotEncoding

```python
# We need to extract the original categories for each encoded column
original_categories = {col: encoder.categories_[i] for i, col in enumerate(cat_columns)}
# Manual creation of feature names
feature_names = []
for i, col in enumerate(cat_columns):
    for category in encoder.categories_[i]:
        feature_names.append(f"{col}_{category}")

encoded_df = pd.DataFrame(encoded_df, columns=feature_names)
pd.set_option('display.max_columns', None)
df.head(10)
```

## Inverse transform of OneHotEncoding
### Can OneHotEncoding be inverse transform?
- No, one-hot encoding does not have a direct inverse transform. One-hot encoding is a technique used to convert categorical variables into a binary vector representation, where each category is represented by a binary value (0 or 1) in a separate column. The purpose of one-hot encoding is to enable machine learning algorithms to effectively interpret and utilize categorical data.

- While it is possible to reverse-engineer the original categorical variable from a one-hot encoded representation, there is no built-in inverse transform function for this purpose. The reverse process involves identifying the column with the value of 1 for each observation and mapping it back to the corresponding category. However, this process is not standardized and requires custom logic to be implemented.

- If you need to revert one-hot encoded data back to its original categorical form, you can use techniques such as finding the column with the maximum value for each observation or creating a mapping dictionary to map the binary vectors back to their respective categories. The approach you choose will depend on the specific requirements of your problem.


# **3. ORDINAL ENCODING**

```python
from sklearn.preprocessing import OrdinalEncoder
df = sns.load_dataset('titanic')
# ordinal encoding the day column using specific order
oe_sex = OrdinalEncoder(categories=[['female', 'male']]) # implicitly specify
oe_class = OrdinalEncoder(categories=[['None','First', 'Second', 'Third']]) # implicitly specify
df['encoded_sex'] = oe_sex.fit_transform(df[['sex']])
df['encoded_class'] = oe_class.fit_transform(df[['class']])
df.head()
```

## Inverse transform of Ordinal Encoding

```python
# Inverse transform data

decoded_sex = oe_sex.inverse_transform(df[['encoded_sex']])
decoded_class = oe_class.inverse_transform(df[['encoded_class']])
df.drop(['encoded_sex', 'encoded_class'], axis=1, inplace=True)
df.head()
```
